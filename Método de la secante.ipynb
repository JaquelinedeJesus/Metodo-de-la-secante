```python
import numpy as np
import matplotlib.pyplot as plt

f= lambda x: np.exp(-x)-x
i=0
Tolerancia=0.000001
Error=1
a=0
b=1
while (Error>Tolerancia and i<=1000):
    xi=b-f(b)*(a-b)/(f(a)-f(b))
    Error=np.abs(f(xi))
    print('i=',i,' a=',a,' b=',b,' xi=',xi,'  f(xi)=',f(xi),' Error=',Error)
    if (f(a)-f(b))/(a-b) >0:
        b=xi
    else:
        a=xi
    i=i+1
if i>1000:
   print('Luego de 1000 intentos no se logró obtener un raíz')
else:
   print('El valor de x, tal que f(x)=0 es:', xi)

x=np.linspace(-3,3,100)
plt.title('Método de la Secante')
plt.plot(x,f(x), label='f(x)', linestyle='dashed')
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.axvline(xi,color='blue', linestyle='dashed',label=f"f(x)=,x={xi:.6f}")
plt.xlabel('xi')
plt.ylabel('f(xi)')
plt.legend(loc='upper right')
plt.grid
plt.show()
```
